```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
from dateutil.relativedelta import relativedelta

warnings.filterwarnings("ignore")
from pathlib import Path
import os

project_dir = Path(os.getcwd()).parent

# Define the path to the raw data folder relative to the project directory
data_dir = project_dir / 'data' / 'data_files' / 'processed_data'

# Load the data
data = pd.read_csv(data_dir / 'deseasonalized_data.csv')

# Convert the date column to datetime and set it as the index
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

# Function to standardize a single column
def z_score(x):
    return (x - x.mean()) / x.std()

normalized = data.copy()

normalized.loc[:, normalized.columns != 'date'] = normalized.loc[:, normalized.columns != 'date'].apply(z_score)

data = normalized

# Select the target variable (GDP)
target_variable = 'China: GDP: Current Prices:PoP GROWTHRATE'

# Check for missing values and handle them if necessary
data[target_variable] = data[target_variable].dropna()

# Split the data into train and test sets
train_data = data[data.index < '2018-06-30'][target_variable].dropna()
test_data = data[data.index >= '2018-06-30'][target_variable].dropna()


# Function to generate artificial testing vintages
def generate_artificial_vintages(data, test_start_date, target_variable):
    test_dates = data[data.index >= test_start_date].index
    vintages = []
    
    for test_date in test_dates:
        train_end_date = test_date - relativedelta(months=1)
        train_data = data[data.index <= train_end_date][target_variable]
        test_value = data.loc[test_date, target_variable]
        vintages.append((train_data, test_date, test_value))
    
    return vintages


# Generate artificial testing vintages
vintages = generate_artificial_vintages(data, '2018-06-30', target_variable)

# Initialize lists to store results
forecasts = []
actuals = []
forecast_dates = []
confidence_intervals = []

# Iterate through vintages
for train_data, test_date, actual_value in vintages:
    # Check for and remove any inf or nan values
    mask = np.isfinite(train_data)
    train_data = train_data[mask]
    
    if len(train_data) == 0:
        print(f"Warning: No valid data for {test_date}")
        continue

    try:
        # Perform ADF test
        result = adfuller(train_data)

        # If the series is not stationary, difference the training data
        if result[1] > 0.05:
            train_data = train_data.diff().dropna()
            d = 1
        else:
            d = 0

        # Define the order (p, d, q) and seasonal order (P, D, Q, s) for SARIMAX model
        p, q = 3, 1
        P, D, Q, s = 1, 1, 1, 4  # Assuming quarterly data, adjust s if different

        # Fit the SARIMAX model
        sarimax_model = SARIMAX(train_data, 
                                order=(p, d, q), 
                                seasonal_order=(P, D, Q, s))
        sarimax_result = sarimax_model.fit()

        # Forecast for the test date
        forecast = sarimax_result.get_forecast(steps=1)
        mean_forecast = forecast.predicted_mean[0]
        conf_int = forecast.conf_int()
        
        forecasts.append(mean_forecast)
        actuals.append(actual_value)
        forecast_dates.append(test_date)
        confidence_intervals.append(conf_int)
    except Exception as e:
        print(f"Error forecasting for {test_date}: {str(e)}")

# Convert forecasts and actuals to numpy arrays, dropping NaN values
forecasts = np.array(forecasts)
actuals = np.array(actuals)
forecast_dates = pd.to_datetime(forecast_dates)
confidence_intervals = np.array(confidence_intervals).squeeze()

# Drop NaN values from actuals and corresponding forecasts, dates, and confidence intervals
mask = ~np.isnan(actuals)
actuals = actuals[mask]
forecasts = forecasts[mask]
forecast_dates = forecast_dates[mask]
confidence_intervals = confidence_intervals[mask]

# Plot the forecast against actual test data for the test period only
plt.figure(figsize=(10, 6))
plt.plot(test_data.index, test_data, label='Actual Test Data')
plt.plot(forecast_dates, forecasts, label='Forecast')
plt.fill_between(forecast_dates, confidence_intervals[:, 0], confidence_intervals[:, 1], color='gray', alpha=0.2, label='95% Confidence Interval')
plt.title('ARIMA Forecast vs Actual (Test Period)')
plt.legend()
plt.show()

# Evaluate the model on the test data
mse = mean_squared_error(actuals, forecasts)
mae = mean_absolute_error(actuals, forecasts)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((actuals - forecasts) / actuals)) * 100

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}%')
```
